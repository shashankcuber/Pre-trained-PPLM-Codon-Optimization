```python
import pandas as pd, textwrap
csv_path = ""
df = pd.read_csv(csv_path)
with open("test_human_long_protbert.fasta", "w") as out:
    idx = 1
    for seq in df["aa_sequence"].fillna("").astype(str).str.strip():
        if not seq:
            continue
        out.write(f">Protein_{idx}_human\n")
        out.write("\n".join(textwrap.wrap(seq, 2000)) + "\n")
        idx += 1
```

```python
# convert fast to csv
fasta_path = ""
csv_path = ""
import csv

def fasta_to_csv(fasta_path, csv_path):
    """
    Convert FASTA file to CSV file with columns: id, sequence
    """
    records = []
    with open(fasta_path, 'r') as f:
        seq_id, seq = None, []
        for line in f:
            line = line.strip()
            if not line:
                continue
            if line.startswith('>'):
                # save previous record
                if seq_id is not None:
                    records.append((seq_id, ''.join(seq)))
                seq_id = line[1:].strip()  # remove '>'
                seq = []
            else:
                seq.append(line)
        # save last record
        if seq_id is not None:
            records.append((seq_id, ''.join(seq)))

    # Write to CSV
    with open(csv_path, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['id', 'sequence'])
        writer.writerows(records)

    print(f"✅ Converted {len(records)} sequences from {fasta_path} → {csv_path}")


# Example usage:
fasta_to_csv(fasta_path, csv_path)

```
